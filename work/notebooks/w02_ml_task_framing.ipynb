## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking since the goal is to rank the pages by how likely they are needed to be refreshed.
A model can estimate the probability that a page is showing decline signals, then it can create a ranked review queue.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

I would predict which pages will be in most need of a refresh with the 'is_declining_label' as the proxy which will created with the column 'trend_direction' which are "down".

It represents whether a page currently shows declining performance signals.



## 3. Success metric

*One metric you can defend. What number means 'good'?*

The success metric for this project is Precision@50. It will measure how many of the top 50 pages recommended by the model are actually pages with decline signals.
A higher Precision@50 means the ranked list is more useful because reviewers spend time on pages that are more likely to actually need to be refreshed.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule could use simple conditions such as old content age, low CTR, and a low engagement.
However, content performance depends on multiple signals interacting together, for exmaple; a newer page with declining visibility may need attention comapared to an older one that has high visibility.
